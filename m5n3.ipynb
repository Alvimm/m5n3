```python
import pandas as pd
df = pd.read_csv('pico_web.csv', sep=';', engine='python', encoding='utf-8')
df
```

```python
new_df = df
new_df['Calories']= new_df['Calories'].fillna(0)
new_df
```

```python
new_df['Date'] = new_df['Date'].fillna('1900/01/01')
new_df

```

```python
new_df['Date'] = pd.to_datetime(new_df['Date'])
```

```python
new_df['Date'] = new_df['Date'].replace('1900/01/01', None)

new_df['Date'] = pd.to_datetime(new_df['Date'])
new_df.tail(10)
```

```python
new_df['Date'] = new_df['Date'].replace('20201226', pd.NaT)

new_df['Date'] = pd.to_datetime(new_df['Date'])
new_df.tail(10)
```

```python
df_cleaned = new_df.dropna(subset=['Date'])
df_cleaned
```
